```python
# Vanila Ensemble
from sklearn.ensemble import VotingClassifier

votingC1  = VotingClassifier(estimators=[('GBC_best', GBC_best), ('XGB_best', XGB_best), ('LGB_best', LGB_best),('LR_best',LR_best)], voting='soft')
pred_1 = votingC1.fit(X_train2_train, y_train_train).predict_proba(X_train2_test)[:,1]
roc_auc_score(y_train_test, votingC1.predict_proba(X_train2_test)[:,1])
```

```python
# Vanila Ensemble
from sklearn.ensemble import VotingClassifier

votingC3  = VotingClassifier(estimators=[('XGB_best_ee', XGB_best_ee), ('LGB_best_ee', LGB_best_ee),('LR_best_ee',LR_best_ee)], voting='soft')
pred_3 = votingC3.fit(X_train2_train, y_train_train).predict_proba(X_train2_test)[:,1]
roc_auc_score(y_train_test, votingC3.predict_proba(X_train2_test)[:,1])
```
